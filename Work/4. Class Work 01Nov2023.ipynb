### 10 Programs of Pandas

#### 1. Creating a dataframe

```python
import pandas as pd
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 30, 22, 35, 28],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami']}

df = pd.DataFrame(data)
df
```

#### 2. Filtering out dataframe rows in which age is greater than 25

```python
age_greater_than_25 = df[df['Age'] > 25]
age_greater_than_25
```

#### 3. Sorting dataframe entities by age in descending order

```python
sorted_df = df.sort_values(by='Age', ascending=False)
sorted_df
```

#### 4. FInding mean age from the pandas dataframe

```python
average_age = df['Age'].mean()
average_age
```

#### 5. creating a new column in dataframe

```python
df['Country'] = ['USA', 'USA', 'USA', 'USA', 'USA']
df
```

#### 6. Filtering out the entries which are equal or greater than 25 and city as New York

```python
selected_rows = df[(df['Age'] >= 25) & (df['City'] == 'New York')]
selected_rows
```

#### 7. Creating a new dataframe from an existing dataframe

```python
new_df = df[['Name', 'Age']]
new_df
```

#### 8. Grouping the dataframe by mean and max

```python
grouped = df.groupby(['City', 'Country']).agg({'Age': ['mean', 'max']})
grouped
```

#### 9. Applying anonymous function in pandas dataframe

```python
df['Age + 5'] = df['Age'].apply(lambda x: x + 5)
df
```

#### 10. Replacing in pandas dataframe

```python
df['City'] = df['City'].replace({'New York': 'NYC', 'Los Angeles': 'LA'})
df
```

#### 11. Making a pivot table from pandas dataframe

```python
pivot_df = df.pivot_table(index='Name', columns='City', values='Age', fill_value=0)
pivot_df
```

### 10 Programs of matplotlib

#### 1. Basic line plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 11]
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Line PLot")
plt.plot(x, y)
```

#### 2. Scatter plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Scatter PLot")
plt.scatter(x, y)
plt.show()
```

#### 3. Bar Chart

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 5]
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Bar chart")
plt.bar(categories, values)
plt.show()
```

#### 4. Histogram

```python
import matplotlib.pyplot as plt
data = [1, 1, 2, 3, 3, 3, 4, 4, 5]
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Histogram")
plt.hist(data, bins=5)
plt.show()
```

#### 5. Pie Chart

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.title("Pie Chart")
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

#### 6. Box Plot

```python
import matplotlib.pyplot as plt
data = [10, 15, 20, 25, 30, 35, 40]
plt.title("BOX PLOT")
plt.boxplot(data)
plt.show()
```

#### 7. heatmap

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(5, 5)
plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.title("HEATMAP")
plt.colorbar()
plt.show()
```

#### 8. 3D Scatter PLot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
plt.show()
```

#### 9. Stacked Bar Chart

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C']
values1 = [10, 15, 20]
values2 = [5, 10, 15]
plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, label='Group 2', bottom=values1)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Stacked Bar chart")
plt.legend()
plt.show()
```

#### 10. Area Plot

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]
plt.title("Area Chart")
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.show()
```

#### 11. Stacked Area Chart

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 1, 3, 2]
y2 = [0, 1, 2, 0, 1]
plt.title("Stacked Area chart")
plt.fill_between(x, y1, color='blue', alpha=0.5)
plt.fill_between(x, y2, color='green', alpha=0.5)
plt.show()
```
