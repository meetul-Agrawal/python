### 17. Dataframe Functions

```python
#-------.apply()-----------------------------
print("Dataframe.apply() :")
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
def double(x):
    return x * 2
result = df.apply(double, axis=0)
print(result)
print()

#----------.assign()---------------------------
print("Dataframe.assign() :")
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
new_df = df.assign(C=10)
print(new_df)
print()

#-----------.astype()----------------------------
print("Dataframe.astype() :")
data = {'A': ['1', '2', '3']}
df = pd.DataFrame(data)
print("Before type casting :",df.dtypes)
df['A'] = df['A'].astype(int)
print("After type casting :",df.dtypes)
print()

#-------------.concat()-------------------------
print("Dataframe.concat() :")
data1 = {'A': [1, 2], 'B': [3, 4]}
data2 = {'A': [5, 6], 'B': [7, 8]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
result = pd.concat([df1, df2], axis=0)
print(result)
print()

#---------.count()----------------------------
print("Dataframe.count() :")
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 2, 3, 4, None],
        'C': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
print(df.count())
print()

#----------.describe()--------------------------
print("Dataframe.describe() :")
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 2, 3, 4, None],
        'C': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
print(df.describe())
print()

#---------.drop_duplicates()---------------------
print("Dataframe.drop_duplicates() :")
data = {'A': [1,2,3,5,9,5,1,5,2,6,7]}
df = pd.DataFrame(data)
print("OG Dataframe :",df)
print("After duplicate deletion :", df.drop_duplicates())
```

```python
#----------.groupby()-------------
print("Dataframe.groupby(): ")
data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 20, 15, 25, 30]
}
df = pd.DataFrame(data)
grouped = df.groupby('Category')['Value'].mean()
print(grouped)
print()

#------------.head()--------------------
print("Dataframe.head(): ")
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'C', 'D'],
        'Value': [10, 20, 15, 25, 30, 56, 84]
}
df = pd.DataFrame(data)
print(df.head())
print()

#------------.hist()--------------------
print("Dataframe.hist(): ")
import pandas as pd
import matplotlib.pyplot as plt
data = {'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65]}
df = pd.DataFrame(data)
df['Age'].hist()
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
print()

#------------.iterrows()--------------------
print("Dataframe.iterrows(): ")
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
for index, row in df.iterrows():
    print("Index :",index,", Name :",row["Name"],", Age :",row["Age"])
print()

#------------.mean()--------------------
print("Dataframe.mean(): ")
data = {'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65]}
df = pd.DataFrame(data)
print(df.mean())
print()

#----------------.melt()----------------------
print("Dataframe.melt(): ")
data = {'ID': [1, 2, 3],
        'Math': [90, 85, 78],
        'Physics': [88, 92, 80]}
df = pd.DataFrame(data)
melted_df = df.melt(id_vars='ID', var_name='Subject', value_name='Score')
print(melted_df)
print()

#----------------.merge()----------------------
print("Dataframe.merge(): ")
data1 = {'ID': [1, 2, 3],
         'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'ID': [2, 3, 4],
         'Grade': ['A', 'B', 'C']}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
merged_df = df1.merge(df2, on='ID')
print(merged_df)
print()

#----------------.pivot_table()----------------------
print("Dataframe.pivot_table(): ")
data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value1': [10, 20, 15, 25, 30],
        'Value2': [100, 200, 150, 250, 300]}
df = pd.DataFrame(data)
pivot_df = df.pivot_table(index='Category', values=['Value1', 'Value2'], aggfunc='mean')
print(pivot_df)
```

```python
#----------------.query()----------------------
print("Dataframe.query(): ")
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
result = df.query('A > 3')
print(result)
print()

#----------------.sample()----------------------
print("Dataframe.sample(): ")
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
print(df.sample(2))
print()

#----------------.shift()----------------------
print("Dataframe.shift(): ")
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
print(df['A'].shift(2))
print()

#----------------.sort_values()----------------------
print("Dataframe.sort_values(): ")
data = {'A': [3, 1, 4, 2, 5],
        'B': ['C', 'A', 'D', 'B', 'E']}
df = pd.DataFrame(data)
sorted_df = df.sort_values(by='A')
print(sorted_df)
print()

#----------------.sum()----------------------
print("Dataframe.sum(): ")
data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
print(df['A'].sum())
print()

#----------------.to_excel()----------------------
print("Dataframe.to_excel(): ")
data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
df.to_excel('data.xlsx', sheet_name='Sheet1', index=False)
print(pd.read_excel('data.xlsx'))
print()

#----------------.transpose()----------------------
print("Dataframe.transpose(): ")
data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}
df = pd.DataFrame(data)
print(df.transpose())
print()

#----------------.where()----------------------
print("Dataframe.where(): ")
data = {'A': [1, 2, 3],
        'B': [4, 5, 6]}
df = pd.DataFrame(data)
print(df.where(df['A'] <= 3))
```

### 19. Data Operations in Pandas

```python
#----------------.isnull()----------------------
print("Dataframe.isnull(): ")
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 2, 3, 4, None]}
df = pd.DataFrame(data)
print(df.isnull())
print()

#----------------.notnull()----------------------
print("Dataframe.notnull(): ")
print(df.notnull())
print()

#----------------.dropna()----------------------
print("Dataframe.dropna(): ")
print(df.dropna())
print()

#----------------.fillna()----------------------
print("Dataframe.fillna(): ")
print(df.fillna(0))
print()

#----------------.fillna()----------------------
print("Dataframe.fillna(): ")
data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
print(df.replace({2: 20, 4: 40}))
print()

#----------------.interpolate()----------------------
print("Dataframe.interpolate(): ")
data = {'A': [1, None, 3, None, 5]}
df = pd.DataFrame(data)
print(df.interpolate())
```

```python
#----------------.lower()----------------------
print("Dataframe.lower(): ")
data = {'A': ["MEETUL", "JALAY"]}
df = pd.DataFrame(data)
print(df["A"].str.lower())
print()

#----------------.upper()----------------------
print("Dataframe.upper(): ")
data = {'A': ["meetul", "jalay"]}
df = pd.DataFrame(data)
print(df["A"].str.upper())
print()

#----------------.strip()----------------------
print("Dataframe.strip(): ")
data = {'A': ["meetul agrawal", "jalay"]}
df = pd.DataFrame(data)
df['B'] = df["A"].str.strip('')
print(df['B']) 
print()

#----------------.replace()----------------------
print("Dataframe.replace: ")
data = {'A': ["meetul agrawal", "jalay"]}
df = pd.DataFrame(data)
print([df['A'].apply(lambda x: [word.replace('meetul agrawal','meetul') for word in x])])
```

### 26. Creating a using sin formula

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
ax.plot(x,np.sin(x))
x=np.arange(0.0,2,0.01)
y1=np.sin(2*np.pi*x)
y2=1.2*np.sin(4*np.pi*x)
fig.ax=plt.subplots(1,sharex=True)
ax.plot(x,y1,x,y2,color='black')
ax.fill_between(x,y1,y2,where=y1<=y2,facecolor='blue',interpolate=True)
ax.fill_between(x,y1,y2,where=y1<=y2,facecolor='red',interpolate=True)
ax.set_title('fill between where')
```
